```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.2, 0.5, 0.25, 0.25])

axes.set_xlim(0,8)
axes.set_ylim(0, 8000)
axes.set_xlabel('A')
axes.set_ylabel('B')
axes.set_title('Power of 4')

axes1.plot(a,b)

axes1.set_xlim(1,2)
axes1.set_ylim(0, 50)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Zoomed In')

axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(2,2), dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```

```python
x = np.arange(0,10)
y = 2 * x
```

## Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(x,y)
axes[1][1].set_title('Title')

fig.suptitle('Figure Level', fontsize=12)

fig.subplots_adjust(wspace=0.5, hspace=0.5)

fig.savefig('new_subplots.png', bbox_inches='tight')
```
